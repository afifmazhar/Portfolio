# Gini Index

## Exploratory Data Analysis

This notebook is a simple exploratory data analysis to view the data before running the regression analysis on it. We check for multicollinearity and view the variables as they change over the years by country.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/final_data.csv')
```

```python
data.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
data
```

```python
data.groupby("Country_code").mean()
```

```python
# plot GINI index vs Year
fig, ax = plt.pyplot.subplots()
sns.lineplot(x=data["Year"],  
                y=data['Gini'],
                hue=data['Country']).set(title=                
                'Gini Index over the Years')
 
plt.pyplot.show()
```

```python
data.groupby("Country_code").std()
```

```python
# plot real GDP index vs Year
fig, ax = plt.pyplot.subplots()
sns.lineplot(x=data["Year"],  
                y=data['real_GDP'],
                hue=data['Country']).set(title=                
                'GDP over the Years')
 
plt.pyplot.show()
```

```python
# plot CPI vs Year
fig, ax = plt.pyplot.subplots()
sns.lineplot(x=data["Year"],  
                y=data['cpi'],
                hue=data['Country']).set(title=                
                'CPI over the Years')
 
plt.pyplot.show()
```

```python
# plot IP vs Year
fig, ax = plt.pyplot.subplots()
sns.lineplot(x=data["Year"],  
                y=data['ip'],
                hue=data['Country']).set(title=                
                'IP over the Years')
 
plt.pyplot.show()
```

```python
# plot Unemployment  vs Year
fig, ax = plt.pyplot.subplots()
sns.lineplot(x=data["Year"],  
                y=data['unrate'],
                hue=data['Country']).set(title=                
                'Unemployment Rate over the Years')
 
plt.pyplot.show()
```

```python
fig, ax = plt.pyplot.subplots(figsize=(16,16)) 
sns.heatmap(data=data.corr(method='spearman'), annot=True, ax=ax, fmt ='.3f')
```
