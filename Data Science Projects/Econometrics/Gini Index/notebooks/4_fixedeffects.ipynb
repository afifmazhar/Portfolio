```python
import pandas as pd
import scipy.stats as st
import statsmodels.api as sm
import statsmodels.formula.api as smf
from matplotlib import pyplot as plt
import seaborn as sns
from linearmodels import PanelOLS
```

```python
data = pd.read_csv('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/final_data.csv', header = 0)
data.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
data.head()
```

```python
unit_col_name='Country'
time_period_col_name='Year'
 
#Create the dummy variables, one for each country
data_dummies = pd.get_dummies(data[unit_col_name])
```

```python
data_w_dummies = data.join(data_dummies)
```

```python
data_w_dummies.head()
```

```python
y = 'Gini'
X = ['real_GDP','cpi','ip','unrate']
unit_names = ['Australia', 'Sweden', 'Germany', 'Spain', 'United Kingdom', 'United States']
exog_X = sm.add_constant(data[X])
exog_X
```

```python
lsdv_expr = y + ' ~ '
i = 0
for X_var_name in X:
    if i > 0:
        lsdv_expr = lsdv_expr + ' + ' + X_var_name
    else:
        lsdv_expr = lsdv_expr + X_var_name
    i = i + 1
for dummy_name in unit_names[:-1]:
    lsdv_expr = lsdv_expr + ' + ' + dummy_name
 
print('Regression expression for OLS with dummies = ' + lsdv_expr)
```

```python
mod = PanelOLS(data.Gini, exog_X, entity_effects = True)

re_res = mod.fit()

print(re_res)
```
