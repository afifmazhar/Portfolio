```python
import pandas as pd
import numpy as np
```

```python
australia = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/Australia_final.xlsx')
sweden = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/Sweden_final.xlsx')
germany = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/Germany_final.xlsx')
spain = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/Spain_final.xlsx')
uk = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/UK_final.xlsx')
usa = pd.read_excel('/Users/afifmazhar/Desktop/Data Science/Data Science Projects/Econometrics/Gini Index/data/USA_final.xlsx')
```

```python
usa.head()
```

```python
data = [australia, sweden, germany, spain, uk, usa]
float_variables = ['Gini','real_GDP','cpi','ip','unrate']
```

```python
# for loop that fills the null values of each column of each excel file
for country in data:
    for column in usa:
        if column in float_variables:
            country[column].fillna(country[column].mean(), inplace = True)
        else:
            pass 
```

```python
# for loop that checks each column of each excel file to determine if null values still exist
for country in data:
    print(country.isna().sum())
```

```python
# join the excel files together to create the panel dataset
final_data = pd.concat(data, ignore_index = True)
```

```python
final_data = final_data[~(final_data['Year'] < '1992-02-01')]
```
